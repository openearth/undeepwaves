```python
import pathlib

import xarray as xr

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_dir = pathlib.Path(r'P:\11207539-001-undeepwaves')
```

```python
runs_dir = data_dir / 'bathy-emodnet-a-runs'
```

```python
run_dir = list(runs_dir.glob('*'))[20]
```

```python
result_file = None
for nc_file  in run_dir.glob('results/*.nc'):
    if '-fr' not in nc_file.name:
        result_file = nc_file
        
file_found = result_file is not None
```

```python
#  open a dataset
ds = xr.open_dataset(result_file)
```

```python
fig, axes = plt.subplots(ncols=4, figsize=(16, 4))
im = axes[0].imshow(ds.variables['botl'][0])

plt.colorbar(im, ax=axes[0])
im = axes[1].imshow(ds.variables['hs'][0])
plt.colorbar(im, ax=axes[1])
im = axes[2].imshow(ds.variables['ssh'][0])
plt.colorbar(im, ax=axes[2])
im = axes[3].imshow(ds.variables['tm01'][0])
plt.colorbar(im, ax=axes[3])
```
